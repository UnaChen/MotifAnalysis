```python
import os

def isExe(f):
    if f.endswith("exe"):
        return True
    else:
        return False

def isMZ(f):
    with open(f, 'rb') as hnd:
        if hnd.read(2) == "MZ":
            return True
        else:
            return False

def isPE(f):
    try:
        with open(f, 'rb') as hnd:
            if hnd.read(2) != b"MZ":
                return False
            hnd.seek(0x3c, 0)

            import struct
            offset = struct.unpack('i', hnd.read(4))[0]
            hnd.seek(offset, 0)
            if hnd.read(4) == b"PE\0\0":
                ## MIKE: for 64 or 32 bit
                machine = struct.unpack("<H", hnd.read(2))[0]
                if machine == 332: # IMAGE_FILE_MACHINE_I386
                    print(f, "IA-32 (32-bit x86)")
                elif machine == 512: #IMAGE_FILE_MACHINE_IA64:
                    print(f, "IA-64 (Itanium)")
                elif machine == 34404: #IMAGE_FILE_MACHINE_AMD64:
                    print(f, "AMD64 (64-bit x86)")
                else:
                    print(f, "Unknown architecture")
                ## end
                return True
            else:
                print(f, "isMZ, not PE.")
                return False
        
    except:
        print(f, "Permission denied")
        return False

def isWin32EXE(f):
    try:
        with open(f, 'rb') as hnd:
            if hnd.read(2) != b"MZ":
                return False
            hnd.seek(0x3c, 0)

            import struct
            offset = struct.unpack('i', hnd.read(4))[0]
            hnd.seek(offset, 0)
            if hnd.read(4) == b"PE\0\0":
                ## MIKE: for 64 or 32 bit
                machine = struct.unpack("<H", hnd.read(2))[0]
                if machine == 332: # IMAGE_FILE_MACHINE_I386
                    #print(f, "IA-32 (32-bit x86)")
                    return True
                elif machine == 512: #IMAGE_FILE_MACHINE_IA64:
                    #print(f, "IA-64 (Itanium)")
                    pass
                elif machine == 34404: #IMAGE_FILE_MACHINE_AMD64:
                    #print(f, "AMD64 (64-bit x86)")
                    pass
                else:
                    #print(f, "Unknown architecture")
                    pass
                ## end
                return False
            else:
                #print(f, "isMZ, not PE.")
                return False
        
    except:
        #print(f, "Permission denied")
        return False
```
